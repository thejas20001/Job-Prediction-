## Logistic Regression


```python
import pandas as pd
```

```python
data = pd.read_csv('score-210418-184515.csv')  
```

```python
data                #labels will be admitted and gmat gpa work will be features
```

```python
X = data.iloc[:,0:3]      #to extract features  # can be written as X = data[['gmat','gpa','work_experience']]
```

```python
X
```

```python
Y = data.iloc[:,3]                 # to extraxtlabels
```

```python
Y = data[['admitted']]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest, ytrain, ytest = train_test_split(X,Y,test_size = 0.01)
```

```python
xtrain
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

### Prediction

```python
test = pd.read_csv(r'stest-210418-184515.csv')
```
